## 2. Exploring Relationships

### Correlations:

```python
# Getting column names of dataframe

us_shoot.columns
```

```python
# Creating a subset excluding id, date, manner_of_death, armed, year, month, day, city, state, threat_level, flee

sub = us_shoot[['age','gender','signs_of_mental_illness','body_camera','race_Asian','race_Black','race_Hispanic','race_Native','race_Other','race_White']]
```

```python
# Checking output for sub

sub
```

```python
# Create a correlation matrix for sub

sub.corr()
```

```python
# Create a correlation heatmap using matplotlib

plt.matshow(sub.corr())
plt.show()
```

```python
# Add labels and legend to heatmap

plt.matshow(sub.corr()) # plot type
plt.xticks(range(sub.shape[1]), sub.columns, fontsize=14, rotation=45) # xaxis labels
plt.yticks(range(sub.shape[1]), sub.columns, fontsize=14) # yaxis labels
cb = plt.colorbar() # add a color legend
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

### Creating a correlation heatmap using seaborn:

```python
# Create a subplot with matplotlib

f,ax = plt.subplots(figsize = (10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above

corr = sns.heatmap(sub.corr(), annot = True, ax = ax) # adding correlation coefficients onto the heatmap
```

```python
# Exporting heatmap 

corr.figure.savefig(os.path.join(path,'Analysis', 'Visualizations', 'corr_heat_map.png'))
```

### Scatterplots
